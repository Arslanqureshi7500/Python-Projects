```python
#1.	Install required module 
import pandas as pd
import requests
import pandas as pd
from bs4 import BeautifulSoup
import datetime as dt
```

```python
#2.	Firstly, request for response to get data access
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
```

```python
"""3.	Then, read url using request module and save in variable response 
(if return answer 200 than you get access of specific website)"""
response = requests.get(url)
response
```

```python
#4.	Extract the html_text.
html_text = response.text
html_text
```

```python
"""5.	Create instance of BeautifulSoup and parse with html_parser 
and save in the variable namely soup."""

soup = BeautifulSoup(html_text, "html.parser")
soup
```

```python
"""
6.	Now, extract the table.
7.	Find out specific table class define with using soup.
find and save it in the variable namely table.
"""

table = soup.find("table", {"class":"wikitable"})
table
```

```python
"""
9.	Firstly, extract all table rows of the table with the help of method find_all 
and pass params “tr” and save it in the variable namely rows.
"""

rows = table("tr")
rows
```

```python
"""
10.	Second, extract all the table th = table headers from the rows. 
Use for loop in rows and find out table header using find_all method params ‘th’ 
and save it in th_lst; and once again using for loop from th_lst;
and find out the headers and it append in the empty headers list 
with Data cleaning i.e. beginner and ending spaces remove.
"""

headers = []
for row in rows:
    th_lst = row.find_all("th")
    for h in th_lst:
        headers.append(h.text.strip())
```

```python
"""
a.	Print the array of headers.
"""
headers
```

```python
"""
b.	Fill the null value index with Rank and Notes respectively.
"""
#add label Rank in the header on index 0
headers[0] = "Rank"
#add label Notes in the header on the last index
headers[-1] = "Notes"
headers
```

```python
"""
11.	Third, extract all the table data = table data from the rows. 
Use for loop in rows and find out table data using find_all method params ‘td’
then save it in variable namely td, 
then create a empty list namely table_data 
then once again using for loop from td and find out the table data,
and it extract with .text attribute then it append in the empty table_data list, 
then all table data list save in the empty list namely final_data,
but before it append we can do the Data cleaning 
i.e. beginner and ending spaces remove.
a.	Print the array of final_data.

"""

final_data = []

for row in rows:
    td = row.find_all("td")
    table_data = []
    for d in td:
        table_data.append(d.text.strip())
    final_data.append(table_data)

final_data
```

```python
"""
12.	Now check if you have any empty list than you can slicing.
"""
final_data = final_data[1:]
final_data
```

```python
"""
13.	And check the first column of data 
i.e. ranking data if any data misplace than set the ranking.
"""
final_data[0][0] = "-"
final_data[1][0] = "1"
final_data[2].insert(0, 2)
final_data
```

```python
"""
14.	Now convert all the data into pandas data frame using pd.DataFrame,
and pass params the data set which extracted and save it in variable namely pop.
"""

pop = pd.DataFrame(final_data)
pop
```

```python
"""
15.	Now, we assign our header list to the datafram using .columns attribute.
"""
pop.columns = headers
pop
```

```python
"""
16.	Data Cleaning:
a.	Convert the population column into integer using for loop and save it
in empty list namely popul and update column with data cleaning.
"""

pop["Population"] = pop["Population"].str.replace(",","").str.replace("%","")

popul = []


for i in pop["Population"]:
    p = int(i)
    popul.append(p)

pop["Population"] = popul
```

```python
"""
d.	Formatting the date.
i.	Firstly, check datetime type and data.
ii.	Formatting the datetime to create a function name convert_date and params date, 
then use if else to check the date length that if length of my date is less than 10 
than pass to try and except, else apply full format with dt.datetime.strptime 
and parse it with date demo and format, and save in the variable namely date 
and return date.
iii.	Standardized the Date column, then apply convert_date function 
to the data frame with selecting date column.
iv.	Then updated the date column.
"""
def conv_date(date):
    if len(date) < 10:
        try:
            date = dt.datetime.strptime(date, "%b %Y")
        except:
            date = dt.datetime.strptime(date, "%Y")
    else:
        date = dt.datetime.strptime(date, "%d %b %Y")
        
    return date
```

```python
pop["Date"] = pop["Date"].apply(conv_date)
```
